###Title: Numpy refresher.

Numpy is a library deals with numbers which are represented in arrays, the array manipulation can be performed using python but using numpy for array calculations and other operations gives cutting edge over memory utilization, performance and other things.

```python
#This cell is only for imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creating one-dimension array and print following things
#shape
#dtype
#dim
onedim_array = np.array([0, 1, 2, 3, 4, 5])
print(f'Shape : {onedim_array.shape}')
print(f'd-type : {onedim_array.dtype}')
print(f'dimension : {onedim_array.ndim}')
```

```python
#creating 3x4 matrix and reshape 2x6
matrix = np.array([[1, 2, 3, 4],[ 5, 9,  6, 4], [7, 5, 6, 8]])
matrix_reshape = matrix.reshape(2, 6)
matrix_transpose = matrix_reshape.transpose()
print(f'matrix : \n{matrix}')
print(f'\nmatrix reshape: \n{matrix_reshape}')
print(f'\nmatrix transpose: \n{matrix_transpose}')
```

```python
#Element-wise vs matrix mult: demonstrate * (Hadamard) vs @ (matmul)

a = np.array([[1, 2], [3, 4]])
b = np.array([[4, 5], [6, 8]])

star_mul = a * b
# a * b will do the position wise multiplication, like [[1*4 2*5] [3*6 4*8]]
print(f'\n a * b : \n{star_mul}')

matmul = a@b
#matmul will do proper matrix multiplication i.e., resultant element will be like follows
#for resultant 0thx0th element - 1*4+2*6
#0x1 - 1*5+2*8
#1x0 - 3*4+4*6
#1x1 - 3*5+4*8

print(f'\n a@b: \n{matmul}')
```
