```python
import mlflow, mlflow.sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load iris data.
x, y = load_iris(return_X_y=True)

#split training data into 80-20 split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f'Train data size: {X_train.shape}, Test data size: {X_test.shape}')
```

### ❓ Why don't we evaluate on the training data?

Evaluating on training data does not make sense because the model has already seen this data during training.  
It may simply memorize patterns instead of learning to generalize.

The actual evaluation should be done on **unseen data** (test dataset) to measure how well the model performs in real-world scenarios.

---

### ❓ What does `random_state` do?

`random_state` is used to make results reproducible.

By setting a fixed `random_state` (seed), the same train-test split will be generated every time you run the code.  
This ensures consistency in experiments and debugging.

---

### ❓ What's the shape of `X_train` if Iris has 150 rows and `test_size = 0.2`?

- Total samples = 150  
- Test size = 20% → 30 samples  
- Train size = 150 - 30 = 120 samples  
X_train.shape = (120, 4)

```python
mlflow.set_experiment("iris-hello-world")
with mlflow.start_run():
    c = 95
    max_iter = 200
    acc = 0
    model = LogisticRegression(C=c,max_iter=max_iter)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy_score = accuracy_score(y_test, y_pred)
    mlflow.log_param("C", c)
    mlflow.log_metric("accuracy", accuracy_score)
    mlflow.sklearn.log_model(model, "model")
    print(f'accuracy_score : {accuracy_score}')
```
