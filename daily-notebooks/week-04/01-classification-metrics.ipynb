### "If I told you accuracy is 95%, why might that be misleading? Invent a scenario with numbers."

> - What i will get when someone says 95% accuracy, so consider you are doing the experiment for 100times, in this model suppose you we are doing spam filter, when you send 100 mails one by one 95times it will correctly whether the mail is spam or not.


Call answers the question: How many of them correctly identified? (actual spam mails are 20, then call gives 10, TPR will be 50% because out of 20 spam mails only 10 of them were identified as spam)
Precision answers the question: How many are correctly classified as positives? (model predicted 10 spam mails out of 20 spam mails and all of identified spam mails are spam only then precision becomes 100%)

100 mails are there...

actual:
20 - spam mails 
80 - not spam mails

model predicted:

10 - spam mails (TP)
10 - not spam (FN)
80 - not spam mails (TN)

TPR:

10 / (10 + 10) = 0.5% (50 %)


Precision :
precision = 10 / (10 + 0) = 100%

```python
y_true = [1, 0, 1, 1, 0, 0, 1, 0, 1, 1]
y_pred = [1, 0, 0, 1, 0, 1, 1, 0, 0, 1]
```

                Predicted spam      Predicted not-spam
Actual spam           4                     2
Actual not-spam       1                     3

TP : 4
FP : 1
TN : 3
FN : 2

TPR = TP / (TP + FN) = 4/6 = 0.67
Precision = TP / (TP+FP) = 4/4+1 = 4/5 = 0.8


```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
tn, fp, fn, tp = confusion_matrix(y_true=y_true, y_pred=y_pred).ravel().tolist()
print(f'tp: {tp}, fp: {fp}, tn: {tn}, fn: {fn} ')

print(classification_report(y_true,y_pred))
```

```python
X, y = make_classification(n_samples=200, n_features=2, random_state=42, n_redundant=0, n_clusters_per_class=1)

model = LogisticRegression(random_state=42)
model.fit(X, y=y)

model.predict(X)

x1_range = np.linspace(
    X[:, 0].min() - 1,
    X[:, 0].max() + 1,
    200
)

x2_range = np.linspace(
    X[:, 1].min() - 1,
    X[:, 1].max() + 1,
    200
)

xx1, xx2 = np.meshgrid(x1_range, x2_range)

grid = np.c_[xx1.ravel(), xx2.ravel()]

probs = model.predict_proba(grid)[:, 1]
print(probs)
```
