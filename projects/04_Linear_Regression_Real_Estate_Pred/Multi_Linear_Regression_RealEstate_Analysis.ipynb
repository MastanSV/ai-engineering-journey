```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

data = {
		    'Size_sqft': [1200, 1500, 1800, 2000, 2200, 2500, 2800, 3000, 3200, 3500],
		    'Bedrooms': [2, 3, 3, 4, 3, 4, 5, 4, 5, 5],
		    'Age_years': [10, 5, 15, 8, 20, 12, 7, 25, 3, 18],
		    'Location_score': [7, 8, 6, 9, 5, 8, 9, 4, 9, 7],
		    'Price_lakhs': [45, 60, 65, 85, 70, 95, 110, 90, 125, 130]
		}


df = pd.DataFrame(data)
df.describe()

X = df[['Bedrooms', 'Age_years', 'Location_score', 'Size_sqft']]
y = df['Price_lakhs']

model = LinearRegression()
model.fit(X,y)
slope_Bedrooms = model.coef_[0]
slope_Age_years = model.coef_[1]
slope_Location_score = model.coef_[2]
slope_Size_sqft = model.coef_[3]

intercept = model.intercept_

print(f'Slope_Bedrooms        : {slope_Bedrooms}')
print(f'Slope__Age_years      : {slope_Age_years}')
print(f'Slope_Location_score  : {slope_Location_score}')
print(f'Slope_Size_sqft       : {slope_Size_sqft}' )

print(f'\nintercept           : {intercept}')
print('\n')
print(df.corr()['Size_sqft'].sort_values(ascending=False))
sns.pairplot(df)


y_pred = model.predict(X)

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)

r2_score = r2_score(y,y_pred)

print(f'MSE         : {mse}')
print(f'RMSE        : {rmse}')
print(f'MAE         : {mae}')
print(f'R2_Score    : {r2_score}')
```
