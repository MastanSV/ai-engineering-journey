```python
# =============================================================================
#                  HOUSE PRICE PREDICTION USING LINEAR REGRESSION
# =============================================================================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# -------------------------- 1. Dataset --------------------------
print("=== Dataset ===")

house_sizes_in_sqft = np.array([800, 1000, 1200, 1500, 1800, 2000, 2500, 3000]).reshape(-1, 1)
house_prices_in_lakhs = np.array([25, 30, 35, 45, 55, 60, 75, 90])

print(f"House Sizes Shape     : {house_sizes_in_sqft.shape}")
print(f"House Prices Shape    : {house_prices_in_lakhs.shape}")
print(f"First 3 sizes (sq ft) : {house_sizes_in_sqft[:3].flatten()}")
print(f"First 3 prices (lakhs): {house_prices_in_lakhs[:3]}")

# -------------------------- 2. Data Visualization --------------------------
print("\n=== Visualizing the Data ===")

plt.figure(figsize=(10, 6))
plt.scatter(house_sizes_in_sqft, house_prices_in_lakhs, color='blue', label='Actual Data')
plt.title('Relationship between House Size and Price')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (in lakhs)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# -------------------------- 3. Model Training --------------------------
print("\n=== Training Linear Regression Model ===")

model = LinearRegression()
model.fit(house_sizes_in_sqft, house_prices_in_lakhs)

# Get model parameters
slope = model.coef_[0]
intercept = model.intercept_
r2_score = model.score(house_sizes_in_sqft, house_prices_in_lakhs)

print(f"Slope (m)     : {slope:.6f}  → ₹{slope*100000:.0f} per sq ft")
print(f"Intercept (c) : {intercept:.4f} lakhs")
print(f"R² Score      : {r2_score:.4f} ({r2_score*100:.2f}%)")

# -------------------------- 4. Predictions & Evaluation --------------------------
print("\n=== Model Performance ===")

# Make predictions
y_pred = model.predict(house_sizes_in_sqft)

# Calculate metrics
mse = mean_squared_error(house_prices_in_lakhs, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(house_prices_in_lakhs, y_pred)

print(f"Mean Squared Error (MSE)  : {mse:.4f}")
print(f"Root Mean Squared Error   : {rmse:.4f} lakhs")
print(f"Mean Absolute Error (MAE) : {mae:.4f} lakhs")

# -------------------------- 5. Final Visualization (Best Fit Line) --------------------------
print("\n=== Final Plot: Actual Data vs Best Fit Line ===")

plt.figure(figsize=(10, 6))
plt.scatter(house_sizes_in_sqft, house_prices_in_lakhs, color='blue', label='Actual Data')
plt.plot(house_sizes_in_sqft, y_pred, color='red', linewidth=2.5, label='Best Fit Line')
plt.title('House Price Prediction using Linear Regression')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price (in lakhs)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# -------------------------- 6. Interpretation --------------------------
print("\n=== Interpretation ===")
print(f"For every additional 1 sq ft, house price increases by approx ₹{slope*100000:.0f} on average.")
print("R² score of 0.9992 indicates a very strong linear relationship between house size and price.")
```
